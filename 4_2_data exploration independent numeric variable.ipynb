# Treating missing values in independent variable

```python
#In independent variable impuattion is prefered over a deletion
#imputation: 1:for continous numeric var we impute by mean and median 2:for categorical var we impute by mode
```

#treating missing values in  numeric variables:

```python
numerical_columns=["No of Bathrooms","Flat Area (in Sqft)","Lot Area (in Sqft)","Lot Area (in Sqft)",
                   "Area of the House from Basement (in Sqft)",
                  "Latitude","Longitude","Living Area after Renovation (in Sqft)"]
```

```python
import numpy as np

from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy="median")
raw_housing_data[numerical_columns]=imputer.fit_transform(raw_housing_data[numerical_columns])
```

# Variable transformation:

# Transforming column zipcode:

```python

#example:
#zipcode 101 may have low price but 102 may have high price as we know the adjacent zipcode may or may have adjacent houses
#so its better to convert this numeric variable to te categorical
```

```python
raw_housing_data["Zipcode"]=raw_housing_data["Zipcode"].astype(object)#type converstion
raw_housing_data["Zipcode"].dtypes
```

# Transforing column "NO of times visited"

```python
#example 2 nd :
#no of times visited is contain data as string none one two etd we change it to numbers
```

```python
raw_housing_data["No of Times Visited"].unique()
```

```python
mapping={"None":"0","Once":"1","Twice":"2","Thrice":"3","Four":"4"}
```

```python
raw_housing_data["No of Times Visited"]=raw_housing_data["No of Times Visited"].map(mapping)
```

```python
raw_housing_data["No of Times Visited"].unique()
```

# transforming Column "Renovated Year"

```python
#deriving two new variables from the existing varibale renovated year:
#1. Ever renovated
#2. year since renovation
```

```python
#1.
raw_housing_data["Ever Renovated"]=np.where(raw_housing_data["Renovated Year"]==0,"No","Yes")
```

```python
raw_housing_data.head()
```

```python
#2
raw_housing_data["Purchase year"]=pd.DatetimeIndex(raw_housing_data["Date House was Sold"]).year
```

```python
raw_housing_data["Years Since Renovation"]=np.where(raw_housing_data["Ever Renovated"]=="Yes",
                                                   abs(raw_housing_data["Purchase year"]-raw_housing_data["Renovated Year"]),0)
```

```python
raw_housing_data.head()
```

```python
#now we have generated the varibale which is actually useful for model building,
#remaining varibales are now been droped using drop function
```

```python
raw_housing_data.drop(columns=["Purchase year","Renovated Year","Date House was Sold"],inplace=True)
```

```python
raw_housing_data.head()
```
