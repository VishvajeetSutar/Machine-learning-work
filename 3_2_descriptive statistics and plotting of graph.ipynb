# Visualization Of Data:

# line graph

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(raw_housing_data['Sale Price'])
```

```python
plt.plot(raw_housing_data["Sale Price"],color="green")
plt.xlabel("record")
plt.ylabel("Sale Price")
plt.title("My First Graph")
plt.show()
```

```python
plt.plot(raw_housing_data["Sale Price"],color="red",marker="o",
         markersize=5,markerfacecolor="blue",linewidth=2, linestyle="dashed")
```

# piechart 

#pie chart mostly used for categorical varible .grouping of data is needed for it.
grouping of data means finding count of record for each unique values of a variable. for this we use groupby function which
groups the categorical or string variable.


```python
raw_housing_data.groupby("Condition of the House")["ID"].count()# pandas also have groupby funcn
```

```python
values=[30,1701,14031,5679,172]
```

```python
condition=["Bad","Excellent","Fair","Good","Okay"]
```

```python
plt.pie(values,labels=condition)
```

```python
plt.bar(condition,values)
```

# Bar plot:


```python
plt.bar(condition,values,color="Orange",linewidth=5,linestyle="dashed")
plt.xlabel("Condition OF The House")
plt.ylabel("Count Of House ")
plt.title("My first Pie chart")
plt.show()
```

# scatter plot:

```python
plt.scatter(x=raw_housing_data["Flat Area (in Sqft)"],y=raw_housing_data["Sale Price"],color="red",marker="o")
plt.xlabel("Flat Area (in Sqft)")
plt.ylabel("Selling Price Of the House")
plt.title("flat area vs selling price")
plt.show()
```

```python
plt.scatter(x=raw_housing_data["Age of House (in Years)"],y=raw_housing_data["Sale Price"],color="brown",marker="o")
plt.xlabel("Age of House (in Years)")
plt.ylabel("Selling Price Of the House")
plt.title("Age of House (in Years) vs selling price")
plt.show()
```
